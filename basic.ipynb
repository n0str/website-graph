```python
from graph_tool.all import *
```

```python
import json
data = json.loads(open("matrix.json").read())
g = Graph()

v_id = g.new_vertex_property("int")

vertices = {int(el): g.add_vertex() for el in data.keys()}
for v1 in data:
    v1_vertex = vertices[int(v1)]
    v_id[v1_vertex] = int(v1)
    for v2 in data[v1]:
        v2_vertex = vertices[int(v2)]
        g.add_edge(v1_vertex, v2_vertex)
g.vertex_properties["v_id"] = v_id
```

```python
sorted_vertices = list(zip(vertices, [t for t in pagerank(g, epsilon=1e-09)]))
sorted_vertices.sort(key=lambda tup: tup[1], reverse=True)
sorted_vertices
```

```python
g.save('graph.dot')
```
